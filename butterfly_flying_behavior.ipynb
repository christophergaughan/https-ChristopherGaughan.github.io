## Here I have found some random data online for Length of Butterfly flights associated with certain of their behavioral traits. Again, Seaborn Makes it easy to see relations. I also show one example of PLOTLY, whicj may not be functional on Github 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load Data

```python
df =  pd.read_excel('Mcinxia_Spain14_Field_FlightDurations.xls')
```

## Obtain Data Headings

```python
df.head(3)
```

## Begin Plotting

```python
ax =sns.scatterplot(x = 'Behavior', y = 'Duration (min:sec)', hue = 'Behavior', data = df)
ax.set_yticklabels(ax.get_xticklabels(),rotation=90)
```

## Seaborn Plots- Pairplot gives all plot relations

```python
sns.pairplot(df, palette="husl")
```

## Seaborn Violin Plots and Scatter Plots 

```python
ax = sns.violinplot(x='Temp_Shade1 (oC)', y='log10 flight duration (s)', data=df, pallette = 'RdBu')
ax.set_xticklabels(ax.get_xticklabels(),rotation=75
                   )
```

```python
sns.scatterplot( x = 'Distance_Flown (m)', y = 'log10 flight duration (s)', data = df, hue='Temp_Shade2 (oC)')
```

```python
sns.scatterplot( x = 'log10 flight duration (s)', y = 'Distance_Flown (m)', data = df)
```

## Seaborn KDE Plot from made-up normalized data

```python
np.random.seed(10)
sns.set(color_codes=True)
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
ax = sns.kdeplot(x)
```

```python
ax = sns.kdeplot(x, shade=True, color="r")
```

```python
ax = sns.kdeplot(x, y)
```

```python
ax = sns.kdeplot(x, y, shade=True)
```

```python
sns.kdeplot(x, y, n_levels=30, cmap="Purples_d")
```
